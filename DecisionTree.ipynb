```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
d1 = {">=9":0, ">=8":1, "<8":2}
d2 = {"yes":0, "no":1}
d3 = {"v.good":0, "good":1, "average":2}
d4 = {"good":0, "moderate":1, "poor":2}
df["cgpa"] = df["cgpa"].map(d1)
df["interact"] = df["interact"].map(d2)
df["pk"] = df["pk"].map(d3)
df["cs"] = df["cs"].map(d4)
df["jo"] = df["jo"].map(d2)
df
```

```python
features = ["cgpa", "interact", "pk", "cs"]

x = df[features]
y = df["jo"]

dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)

tree.plot_tree(dtree, feature_names=features)
```
