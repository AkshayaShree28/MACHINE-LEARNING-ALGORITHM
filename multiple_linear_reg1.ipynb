```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
data = pd.read_csv("DSS4.csv")
data
```

```python
df=data
y=data.X5
d=df.drop(["X5"],axis=1)
d.insert(0,"Bias",1)
d
```

```python
res1=np.linalg.inv(np.matmul(d.T.to_numpy(),d.to_numpy()))
res1
```

```python
res2=np.matmul(d.T.to_numpy(),y.to_numpy())
res2
```

```python
res3=np.matmul(res1,res2)
res3
```

```python
y_pred=res3[0]+res3[1]*d.X1+res3[2]*d.X2+res3[3]*d.X3+res3[4]*d.X4
y_pred
```

```python
y_bar=sum(y)/len(y)
y_bar
```

```python
s1=(y_pred-y_bar)*(y_pred-y_bar)
```

```python
s2=(y-y_bar)*(y-y_bar)
```

```python
r2=sum(s1)/sum(s2)
r2
```

```python
MS=(y-y_pred)**2
MSE=sum(MS)/(len(y))
RMSE=math.sqrt(MSE)
```
